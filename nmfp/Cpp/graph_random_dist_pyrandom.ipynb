```python
#https://docs.python.org/3/library/random.html

from ROOT import TCanvas
from ROOT import TH1F, TLine
from ROOT import gStyle
from array import array
import math
import numpy as np
import random

c1 = TCanvas( 'c1', 'Random Number Generator', 0, 0, 1200, 1000 )
c1.SetGridx()
c1.SetGridy()
c1.GetFrame().SetFillColor( 21 )
c1.GetFrame().SetBorderMode(-1 )
c1.GetFrame().SetBorderSize( 5 )
```

```python
npoints = 10000000
iseed = 1
nbins = 100

random.seed(iseed)

h1 = TH1F("h1","Random Number Generation",nbins,0.0,1.0)

for i in range(0,npoints):
    r = random.random()
    if (i<10): print (i,r)
    #print (i,r)
    h1.Fill(r,1.0*nbins/npoints) 
    
```

```python
error = 3.0/math.sqrt(npoints/nbins)
print ("Three-sigma Error = %f" % error)
topline = TLine(0,1+error,1,1+error)
bottomline = TLine(0,1-error,1,1-error)
topline.SetLineColor(2)
bottomline.SetLineColor(2)

h1.GetYaxis().SetRangeUser(1-2.0*error, 1.0+2.0*error);

h1.Draw("HIST")
topline.Draw("SAME")
bottomline.Draw("SAME")
c1.Draw()
```
