# Coin Systems

Most coin systemx are designed to mimimize the $number$ of coins needed to make change for any amount.  For example, the coin system for the United States is {1, 5, 10, 25, 50, 100}.  Thus, it is a six-coin system.  Of course, the U.S. also has a one-dollar bill, and so the dollar coin is rarely used.  The same is true for the 50 cent coin, and so in practice, the U.S. has a four-coin system that is used for amounts less than one dollar.

The coin system for the Euro is {1, 2, 5, 10, 20, 50, 100, 200}. Thus, it is an eight-coin system.  The smallest Euro note is 5 Euros, and so coind are used in practice for amounts less than 5 Euros.

```python
# The coin system for the United States

us_coins = [25, 10, 5, 1]

# The coin system for the Euro
euro_coins = [200, 100, 50, 20, 10, 5, 2, 1]

n_coins_us = []
n_coins_euro = []
cost_array = []

for cost in range(1,500):
    cost_array.append(cost)
    
    n_coins = 0
    remaining = cost%100
    #print ("Starting cost", cost)
    for coin in us_coins:
        while remaining >= coin:
            n_coins += 1
            remaining -= coin
            #print("Coin", coin, "remaining", remaining)
    print("Cost", cost, " in U.S. requires", n_coins, "coins")
    n_coins_us.append(n_coins)
    
    n_coins = 0
    remaining = cost
    #print ("Starting cost", cost)
    for coin in euro_coins:
        while remaining >= coin:
            n_coins += 1
            remaining -= coin
            #print("Coin", coin, "remaining", remaining)
    print("Cost", cost, "in Europe requires", n_coins, "coins")
    n_coins_euro.append(n_coins)
```

```python
import matplotlib.pyplot as plt
plt.plot(cost_array, n_coins_us, 'o', label="U.S.")
plt.plot(cost_array, n_coins_euro, 'o', label="Euro")
plt.xlabel("Cost")
plt.ylabel("Number of Coins")
plt.legend()
plt.show()
```

```python
import numpy as np
print("The U.S. coin system requires %.2f coins, on average." % np.mean(n_coins_us))
print("The Euro coin system requires %.2f coins, on average." % np.mean(n_coins_euro))
```

# It's not so much the number of coins, it's the mass of the coins

```python
mass_us = [3.11, 5, 2.268, 5.670]
mass_euro = [2.30, 3.06, 3.92, 4.10, 5.74, 7.8, 7.5, 8.5]

mass_us.reverse()
mass_euro.reverse()

print(mass_us)

n_coins_us = []
n_coins_euro = []
cost_array = []

for cost in range(1,500):
    cost_array.append(cost)
    
    n_coins = 0
    remaining = cost%100
    #print ("Starting cost", cost)
    for i in range(len(us_coins)):
        while remaining >= us_coins[i]:
            n_coins += mass_us[i]
            remaining -= us_coins[i]
            #print("Coin", coin, "remaining", remaining)
    print("Cost", cost, " in U.S. requires", n_coins, "grams")
    n_coins_us.append(n_coins)
    
    n_coins = 0
    remaining = cost
    #print ("Starting cost", cost)
    for i in range(len(euro_coins)):
        while remaining >= euro_coins[i]:
            n_coins += mass_euro[i]
            remaining -= euro_coins[i]
            #print("Coin", coin, "remaining", remaining)
    print("Cost", cost, "in Europe requires", n_coins, "grams")
    n_coins_euro.append(n_coins)
```

```python
import matplotlib.pyplot as plt
plt.plot(cost_array, n_coins_us, 'o', label="U.S.")
plt.plot(cost_array, n_coins_euro, 'o', label="Euro")
plt.xlabel("Cost")
plt.ylabel("Number of Coins")
plt.legend()
plt.show()
```

```python
import numpy as np
print("The U.S. coin system requires %.2f grams, on average." % np.mean(n_coins_us))
print("The Euro coin system requires %.2f grams, on average." % np.mean(n_coins_euro))
```

# Is there a better choice?

In principle, we might be able to come up with an even better choice for the U.S. System. We have to have a 1 cent coin, clearly.  However, perhaps we could replace the other three coins (5, 10, 25) with a different set of three coins.  We could try to minimize the number of coins needed to make change for any amount. 



```python
mean_coins = []

min_mean_coins = 999999999999

for coin2 in range(2,11):
    for coin3 in range(coin2+1,18):
        for coin4 in range(coin3+1,100):
            
            us_coins = [coin4, coin3, coin2, 1]
            
            n_coins_us = []
            cost_array = []

            for cost in range(1,500):
                cost_array.append(cost)
    
                n_coins = 0
                remaining = cost%100
                #print ("Starting cost", cost)
                for coin in us_coins:
                    while remaining >= coin:
                        n_coins += 1
                        remaining -= coin
                        #print("Coin", coin, "remaining", remaining)
                #print("Cost", cost, " in U.S. requires", n_coins, "coins")
                n_coins_us.append(n_coins) 
            
            if np.mean(n_coins_us) < min_mean_coins:
                min_mean_coins = np.mean(n_coins_us)
                print("New minimum mean coins", min_mean_coins)
                print("New coin system", us_coins)
            mean_coins.append(np.mean(n_coins_us))
```

```python
import matplotlib.pyplot as plt
plt.plot(cost_array, n_coins_us, 'o', label="U.S.")
plt.xlabel("Cost")
plt.ylabel("Number of Coins")
plt.legend()
plt.show()
```

```python
iteration = [i+1 for i in range(len(mean_coins))]
```

```python
import matplotlib.pyplot as plt
plt.plot(iteration, mean_coins, '.', label="U.S.")
plt.xlabel("Iteration")
plt.ylabel("Mean Number of Coins")
plt.legend()
```
