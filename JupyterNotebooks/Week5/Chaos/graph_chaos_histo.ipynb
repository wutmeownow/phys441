```python
from array import array
import matplotlib.pyplot as plt
```

```python
def growthfunction(xval,par):
    return par[0]*xval*(1.0-xval)
```

```python
lambda_start = 3.99999999
npoints = 100000
xstart = 0.5
iflag = 1
```

```python
lmbda = array('d')
xvalue = array('d')

lmbda.append(lambda_start)

n = npoints
population = []
population.append(xstart)
for i in range(1,n):
    year=i
    x = growthfunction(xstart,lmbda)
    population.append(x)
    xstart = x
```

```python
# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 8))

# Set the grid lines
ax.grid(True)

# Set the background color
ax.set_facecolor('#F0F0F0')

# Set the border properties
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set the canvas size
ax.set_title('Simple Chaos Example')

# Set the line and marker properties
line_color = 'r'
line_width = 2
marker_color = 'b'
marker_style = 'o'
marker_size = 5

# Create the plot
plt.hist(population, bins=50)

# Set the plot title and axis labels
plt.title("Growth Function: lambda*x*(1-x)")
plt.xlabel("Population")
plt.ylabel("Frequency")
#plt.yscale("log")
```
