```python
import matplotlib.pyplot as plt
import numpy as np

# Logistic equation function
def logistic_equation(x, r):
    return r * x * (1 - x)

# Simulation parameters
num_iterations = 100  # Number of iterations
r_values = [2.5, 3.5, 3.9, 4.0]  # Different r values to investigate

# Create subplots with one row and the number of r values columns
num_cols = len(r_values)
fig, axs = plt.subplots(1, num_cols, figsize=(16, 4))

# Iterate over different r values
for i, r in enumerate(r_values):
    # Generate initial x values
    x = np.linspace(0, 1, num=1000)

    # Iterate the logistic equation for each x value
    for _ in range(num_iterations):
        x = logistic_equation(x, r)

    # Plot the final results in the corresponding subplot
    axs[i].plot(np.linspace(0, 1, num=1000), x)
    axs[i].set_title(f'r = {r}')

    # Set plot axis labels
    if i == 0:
        axs[i].set_ylabel('Population')
    axs[i].set_xlabel('x')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
