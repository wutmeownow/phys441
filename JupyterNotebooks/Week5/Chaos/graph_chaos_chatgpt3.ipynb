```python
import matplotlib.pyplot as plt
import numpy as np

# Logistic equation function
def logistic_equation(x, r):
    return r * x * (1 - x)

# Simulation parameters
num_iterations = 1000  # Number of iterations
num_r_values = 100  # Number of r values to investigate
r_values = np.linspace(3.4, 3.95, num=num_r_values)  # Different r values to investigate

# Initialize arrays to store equilibrium population values and corresponding r values
equilibrium_populations = []
r_plot_values = []

# Iterate over different r values
for r in r_values:
    # Generate initial x value
    x = 0.3

    # Iterate the logistic equation until convergence
    for _ in range(num_iterations):
        x_new = logistic_equation(x, r)
        if np.abs(x_new - x) < 1e-2:
            equilibrium_populations.append(x_new)
            r_plot_values.append(r)
            break
        x = x_new

# Plot equilibrium population as a function of r value
plt.plot(r_plot_values, equilibrium_populations, 'o', markersize=2)
plt.title('Equilibrium Population as a function of r')
plt.xlabel('r')
plt.ylabel('Equilibrium Population')

# Show the plot
plt.show()
```
